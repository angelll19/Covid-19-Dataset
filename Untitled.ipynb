We realize that the formula to count  Growth Factor of New Cases is based on 
Today New Cases / Yesterday New Cases 

Note : it must be ordered based on location, date

for 
- Today New Cases =0 and Yesterday New Cases = 0, the growth = 1.00 , which indicates no changes
- Today New Cases = 0 and Yesterday New Cases >0 , the growth =0
- Today New Cases >0 and Yesterday New Cases= 0 => Creates NaN
  even if we fill in, the growth will be unexpectedly high number, so we will non fill that

```python
# is it they hv high new cases bcs of the high population?
data_province = data[data['Location'] != 'Indonesia']
Population_cases=data_province.groupby(['Location', 'Population'])['New Cases'].sum().reset_index()
Population_cases['New Cases'] = Population_cases['New Cases'].astype(int)
Population_cases['Population'] = Population_cases['Population'].astype(int)

```

```python
Population_cases ['Proportion'] = Population_cases['New Cases']/Population_cases['Population']
Population_cases.sort_values(by='Proportion', ascending=False).head(5)
#Jakarta hv the largest population as well as total cases.
# The proportion of total new cases is 13.02% which is doubled of the other region
# Smaller province like Kalimantan Utara only has 648K population but 7% of them infected COVID-19
```

```python
plt.figure(figsize= (10,6))
sns.scatterplot(data=Population_cases,x= 'Population', y='Proportion', hue= 'Location', legend=False)
plt.title('Population vs Total Cases Proportion by Location')
outlier = Population_cases[Population_cases['Location'].isin(['DKI Jakarta','Jawa Tengah', 'Jawa Timur', 'Jawa Barat'])]
for i in range(len(outlier)):
    plt.text(outlier['Population'].iloc[i],
             outlier['Proportion'].iloc[i],
             outlier['Location'].iloc[i],
             fontsize=8, ha='right')
plt.show()

#DKI Jakarta stands out significantly with the highest proportion of cases despite not having the highest population
#Jawa Barat, Jawa Timur, and Jawa Tengah have high population values but relatively low proportions. These areas have lower cases per capita
```

```python
latest_data=data[data['Location']!='Indonesia'].sort_values('Date').drop_duplicates(subset='Location', keep='last')[['Location', 'Date','Total Cases' ,'Total Deaths','Total Recovered', 'Total Active Cases','Island'
]]
latest_data['Death Proportion'] = latest_data['Total Deaths']/latest_data['Total Cases']
latest_data=latest_data.sort_values(by='Death Proportion', ascending=True)
#Drop duplicated rows based on Location only
#Keep only the last row for each location (which is now the row with the latest date). This is bcs the latest date hv the total cases which is sum of all new cases, etc
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x='Total Cases', y='Total Deaths',data=latest_data, hue= 'Island')
for x in range(len(latest_data[latest_data['Island']=='Jawa'])):
    location= latest_data[latest_data['Island']=='Jawa']['Location'].iloc[x]
    proportion= latest_data[latest_data['Island']=='Jawa']['Death Proportion'].iloc[x]
    plt.text(latest_data[latest_data['Island']=='Jawa']['Total Cases'].iloc[x],
             latest_data[latest_data['Island']=='Jawa']['Total Deaths'].iloc[x],
             f"{location}:{proportion:.2%}")
plt.gca().xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.title('Total Cases Vs Total Deaths by Location')
plt.show()

#On the chart, Jawa Timur and Jawa Tengah had fewer total cases than DKI Jakarta and Jawa Barat, but their total deaths were higher.
#In Jawa Tengah and Jawa Timur, approximately 5% of confirmed COVID-19 cases led to death, indicating a relatively high case fatality rate compared to other regions.
#This could mean that even though fewer people got sick in those areas, more of them died, possibly because the healthcare system wasn't as strong.
#Meanwhile, DKI Jakarta and Jawa Barat had more total cases (maybe because of big population or being busy cities), but fewer people died compared to how many got sick. This might be because they had better hospitals, more doctors, or better handling of the pandemic.

```

```python
#is high new cases = bad?
#Do high new cases mean more people are dying
mean_new_cases_popul = data.groupby(['Location']).agg({'New Cases per Million': 'median',
                                                                              'New Deaths per Million' : 'median'}).reset_index()
sns.scatterplot(x='New Cases per Million',y= 'New Deaths per Million', data= mean_new_cases_popul)
some_areas= mean_new_cases_popul[mean_new_cases_popul['Location'].isin(['DKI Jakarta', 'Kalimantan Timur','Jawa Timur','Kalimantan Utara'])]
for i in range(len(some_areas)):
    plt.text(some_areas['New Cases per Million'].iloc[i],
             some_areas['New Deaths per Million'].iloc[i],
             some_areas['Location'].iloc[i], fontsize= 4)


plt.title('New Cases vs New Deaths per Million by Location')
plt.show()
#On a typical day with new cases, Jakarta had around 70 new cases per 1 million people.
# In Jakarta, high new cases lead to high new deaths
# in Kalimantan Utara, even though there's high new cases, the new deaths are low. 
# Kalimantan Timur had high new deaths per million even though the number of new cases are similar to Kalimantan Utara
#This happenned in Jawa Timur as well, where there's high new deaths per million
```

```python
#is high new cases = bad?
mean_new_cases_popul = data.groupby(['Location']).agg({'Total Cases per Million': 'median',
                                                                              'Total Deaths per Million' : 'median'}).reset_index()
sns.scatterplot(x='Total Cases per Million',y= 'Total Deaths per Million', data= mean_new_cases_popul)
some_areas= mean_new_cases_popul[mean_new_cases_popul['Location'].isin(['DKI Jakarta', 'Kalimantan Timur','Jawa Timur','Kalimantan Utara'])]
for i in range(len(some_areas)):
    plt.text(some_areas['Total Cases per Million'].iloc[i],
             some_areas['Total Deaths per Million'].iloc[i],
             some_areas['Location'].iloc[i], fontsize= 4)


plt.title('New Cases vs New Deaths per Million by Location')
plt.show()
#On a typical day with new cases, Jakarta had around 70 new cases per 1 million people.
# In Jakarta, high new cases lead to high new deaths
# in Kalimantan Utara, even though there's high new cases, the new deaths are low. 
# Kalimantan Timur had high new deaths per million even though the number of new cases are similar to Kalimantan Utara
#This happenned in Jawa Timur as well, where there's high new deaths per million
```

```python
yearly_data= data.groupby(['year']).agg({'Total Cases':'sum',
                            'Total Deaths' : 'sum',
                            'Total Recovered' : 'sum',
                            'Total Active Cases':'sum'}).reset_index()
yearly_data ['Death'] = yearly_data['Total Deaths']/ yearly_data['Total Cases']
yearly_data ['Recover'] = yearly_data['Total Recovered']/ yearly_data['Total Cases']
yearly_data ['Active'] = yearly_data['Total Active Cases']/ yearly_data['Total Cases']
```

```python
fig, axes= plt.subplots(1,3,figsize=(20,6))
ax= axes.flatten()
for i,x in enumerate (['Death','Recover','Active']):
    ax[i].pie(yearly_data[x], labels= yearly_data['year'], autopct= '%.1f%%', colors=['pink','lightblue','lightgreen'])
    ax[i].legend()
    ax[i].set_title(f"Proportion of {x} by Year")

# Year by Year, there's less people death, more people recovery, and lesser active case
# The proportion of Death dropped from 38.8% to 28.3% in 2020 to 2022. 
```
